```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ')
```

```python
Rank = []
Name = []
Artist = []
Date = []
Views = []
```

```python
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]"):
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("-")
        
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")
        
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
        
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
        Date.append(i.text)
except NoSuchElementException:
    Date.append("-")
        
try:
    for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")
```

```python
Wiki = pd.DataFrame({})
Wiki['Rank'] = Rank
Wiki['Name'] = Name
Wiki['Uploader'] = Artist
Wiki['Upload Date'] = Date
Wiki['Views (Billions)'] = Views
```

```python
#Scraping title of repository
try:
    Repository_title_tag=driver.find_elements(By.XPATH,'//span[@class="text-normal"]')
    for i in Repository_title_tag:
        Repository_title.append(i.text)
except NoSuchElementException:
    Repository_title.append('-')
    
#Scraping description of Repository
try:
    description=driver.find_elements(By.XPATH,'//article[@class="Box-row"]/p')
    for i in description:
        Repository_description.append(i.text)
except NoSuchElementException:
    Repository_description.append('-')

for i in [Repository_title_tag,description]:
    for j in i:
        if i ==Repository_title_tag:
            Repository_title.append(j.text)
            urls.append(j.get_attribute('href'))
        if i==description:
            Repository_description.append(j.text)
    

    
```

```python
contributors_count=driver.find_elements(By.XPATH,'//a[@class="Link Link--muted d-inline-block mr-3"]')
for i in contributors_count:
    Contributors_count.append(i.text)
```

```python
language_used=driver.find_elements(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]')
for i in language_used:
    Language_used.append(i.text)
```

```python
table = driver.find_element(By.CLASS_NAME, "in-article")
```

```python
Book_name = []
Author_name = []
Volumes_sold = []
Publisher = []
Genre = []


# scraping book names data
for i in driver.find_elements(By.XPATH, "//tbody//tr//td[2]"):
    Book_name.append(i.text)

    
# scraping author names data
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[3]"):
    try:
        if i.text == '0' : raise NoSuchElementException
        Author_name.append(i.text)
    except NoSuchElementException:
        Author_name.append('-')
time.sleep(3)


# scraping data of volumes sold
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[4]"):
    Volumes_sold.append(i.text)
    
    
# scraping data of publisher names
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[5]"):
    Publisher.append(i.text)
    
    
# scraping  data of genre
for i in driver.find_elements(By.XPATH,"//tbody//tr//td[6]"):
    Genre.append(i.text)    
    
```

7. Scrape the details most watched tv series of all time from imdb.com. 
Url = https://www.imdb.com/list/ls095964455/ You 
have to find the following details: 
A) Name 
B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes 

```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get('https://www.imdb.com/list/ls095964455/')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
series_list = soup.find_all("div", class_="lister-item-content")

for series in series_list:
    
    name = series.h3.a.text.strip()
    
    year_span = series.find("span", class_="lister-item-year").text.strip("()")
    
    genre = series.find("span", class_="genre").text.strip()
    
    runtime = series.find("span", class_="runtime").text.strip()
    
    rating = series.find("span", class_="ipl-rating-star__rating").text.strip()

    votes = series.find("span", attrs={"name": "nv"})["data-value"]
    
    print("Name:", name)
    print("Year Span:", year_span)
    print("Genre:", genre)
    print("Run Time:", runtime)
    print("Rating:", rating)
    print("Votes:", votes)
    print("========================")
```
